```python
users["Age"] = users["Age"].fillna(-1).astype(int)
```

```python
def fitness_func(solution, solution_idx):
    # Get the recommended books for all users in the population
    rec_books = books.loc[books['ISBN'].isin(solution.flatten()), :]

    # Calculate the relevance of the recommended books to each user's age and location
    age_mask = rec_books['year'].values.reshape((-1, 1)) >= users.loc[int(solution_idx), 'Age'].values.reshape((1, -1))
    loc_mask = rec_books['publisher'].str.contains(users.loc[int(solution_idx), 'Location'].values.reshape((1, -1)), na=False).values
    user_masks = age_mask & loc_mask
    rec_books = rec_books[user_masks]

    # Calculate the average rating of the recommended books for each user
    rec_books = pd.merge(rec_books, ratings, on='ISBN', how='left')
    rec_books = rec_books.groupby(['ISBN', 'User-ID'])['Book-Rating'].mean().reset_index()
    rec_books = rec_books.groupby('User-ID')['Book-Rating'].mean().reset_index()

    # Calculate the fitness as the average rating of the recommended books
    fitness = rec_books['Book-Rating'].values

    return fitness
```

```python
def solution_encoding(solution_idx):
    # Generate a random set of recommended books for each user in the population
    solutions = []
    for i in range(solution_idx.shape[0]):
        rec_books = set()
        while len(rec_books) < 10:
            rec_books.add(np.random.choice(books['ISBN'].unique()))
        solutions.append(list(rec_books))
    return solutions
```

```python
# Extract the ISBN column
isbn_col = books['ISBN']

# Create a gene_space list from the ISBN column
gene_space = isbn_col.tolist()
```

```python
type(gene_space)
```

```python
import numpy as np
gene_space = np.arange(0, 10)  # for values from 0 to 9

# Set the number of solutions per population
sol_per_pop = 50
 
ga = pygad.GA(num_generations=100, 
              num_parents_mating=10, 
              fitness_func=fitness_func, 
              sol_per_pop=sol_per_pop, 
              num_genes=10, 
              gene_type=int, 
              gene_space=gene_space)

# Initialize the population
low = 0
high = 9
allow_duplicate_genes = False
mutation_by_replacement = True
gene_type = int
ga.initialize_population(low=low, high=high, allow_duplicate_genes=allow_duplicate_genes, mutation_by_replacement=mutation_by_replacement, gene_type=gene_type)
# Run the GA with solution encoding
ga.run()

# Print the best solution and its fitness value for each user
for i in range(users.shape[0]):
    print(f"Best solution for user {i}: {ga.best_solutions_fitness[i]}")
    print(f"Fitness value for user {i}: {ga.best_solutions_fitness[i]}")
```
