```python

import pandas as pd
import random

# define the fitness function
def fitness(individual, books):
    score = 0
    for i in range(len(individual)):
        if individual[i] == 1:
            score += books.iloc[i]['Price']
    return score

# define the selection function
def selection(population, books):
    # calculate the fitness of each individual
    scores = [fitness(individual, books) for individual in population]
    
    # select the best individuals
    selected = []
    for i in range(2):
        index = scores.index(max(scores))
        selected.append(population.pop(index))
        scores.pop(index)
    
    return selected

# define the crossover function
def crossover(parents):
    # select a random crossover point
    point = random.randint(1, len(parents[0])-1)
    
    # create the offspring
    offspring = []
    for i in range(2):
        offspring.append(parents[i][:point] + parents[(i+1)%2][point:])
    
    return offspring

# define the mutation function
def mutation(offspring):
    # select a random mutation point
    point = random.randint(0, len(offspring[0])-1)
    
    # mutate the offspring
    for individual in offspring:
        individual[point] = 1 - individual[point]
    
# define the genetic algorithm function
def genetic_algorithm(books, population_size=10, generations=100):
    
     # create an initial population of random individuals 
     population = [[random.randint(0, 1) for _ in range(len(books))] for _ in range(population_size)]
     
     # run the genetic algorithm 
     for _ in range(generations): 
         # perform selection 
         parents = selection(population, books) 
         
         # perform crossover 
         offspring = crossover(parents) 
         
         # perform mutation 
         mutation(offspring) 
         
         # add the offspring to the population 
         population.extend(offspring) 
    
     # return the best individual from the final population 
     return max(population,key=lambda x:fitness(x,books))

# example usage of genetic_algorithm function with data from CSV file

books=pd.read_csv('/kaggle/input/amazonDataSet/amazon.csv')

best_individual=genetic_algorithm(books)

recommended_books=[book['Name'] for i , book in books.iterrows() if best_individual[i]==1]

print('Recommended Books:',recommended_books)
```
